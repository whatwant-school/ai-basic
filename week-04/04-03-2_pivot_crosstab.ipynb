```python
import pandas as pd
import dateutil
```

### pivot table

```python
!wget https://www.shanelynn.ie/wp-content/uploads/2015/06/phone_data.csv
```

```python
df_phone = pd.read_csv("phone_data.csv")
df_phone['date'] = df_phone['date'].apply(dateutil.parser.parse, dayfirst=True)
df_phone.head()
```

```python
df_phone.pivot_table(["duration"],
                     index=[df_phone.month,df_phone.item], 
                     columns=df_phone.network, aggfunc="sum", fill_value=0)
```

### crosstab

```python
!wget https://raw.githubusercontent.com/TEAMLAB-Lecture/AI-python-connect/master/codes/ch_2/4/data/movie_rating.csv
```

```python
df_movie = pd.read_csv("movie_rating.csv")
df_movie.head()
```

```python
df_movie.pivot_table(["rating"],
                     index=df_movie.critic,
                     columns=df_movie.title,
                     aggfunc="sum",
                     fill_value=0)
```

```python
pd.crosstab(index=df_movie.critic,
            columns=df_movie.title,
            values=df_movie.rating,
            aggfunc="first").fillna(0)
```

```python
df_movie.groupby(["critic","title"]).agg({"rating":"first"}).unstack().fillna(0)
```
